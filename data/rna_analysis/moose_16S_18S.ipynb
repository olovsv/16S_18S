```python
MAGs_names = ['Domain','Phylum','Class','Order','Family','Genus','Species','Subspecies']
in_file_MAGs = pd.read_csv("bins_taxa_all_99.txt",sep='\s+',engine='python',names=MAGs_names)

MAGs_phylum = pd.DataFrame(list(in_file_MAGs.Phylum),columns=['Phylum'])
MAGs_phylum['Phylum'] = MAGs_phylum['Phylum'].str.replace('_',' ')
```

```python
MAGs_phylum = pd.Series(MAGs_phylum.Phylum)
MAGs_phylum = MAGs_phylum.value_counts()
```

```python
rna2save = {}
for c in all_rRNA_bact_arch_phylum.columns:
    Series2Save = all_rRNA_bact_arch_phylum[c].value_counts()
    rna2save[c]= Series2Save
all_rRNA_bact_arch_phylum = pd.DataFrame.from_dict(rna2save)
all_rRNA_bact_arch_phylum['all samples'] = all_rRNA_bact_arch_phylum.sum(axis=1)
all_rRNA_bact_arch_phylum['MAGs'] = MAGs_phylum
```

```python
rna2save = {}
Series2Save = None
for c in all_rRNA_eukar_phylum.columns:
    Series2Save = all_rRNA_eukar_phylum[c].value_counts()
    rna2save[c]= Series2Save
all_rRNA_eukar_phylum = pd.DataFrame.from_dict(rna2save)
all_rRNA_eukar_phylum['all samples'] = all_rRNA_eukar_phylum.sum(axis=1)
```

```python
all_rRNA_bact_arch_phylum = all_rRNA_bact_arch_phylum.fillna(value=np.nan)
all_rRNA_bact_arch_phylum = all_rRNA_bact_arch_phylum.dropna(axis=0, how='all')
all_rRNA_bact_arch_phylum = all_rRNA_bact_arch_phylum.fillna(0)
all_rRNA_bact_arch_phylum
```

```python
all_rRNA_eukar_phylum = all_rRNA_eukar_phylum.fillna(value=np.nan)
all_rRNA_eukar_phylum = all_rRNA_eukar_phylum.dropna(axis=0, how='all')
all_rRNA_eukar_phylum = all_rRNA_eukar_phylum.fillna(0)
all_rRNA_eukar_phylum
```

```python
all_rRNA_bact_arch_phylum = (all_rRNA_bact_arch_phylum / all_rRNA_bact_arch_phylum.sum())
all_rRNA_bact_arch_phylum = all_rRNA_bact_arch_phylum[all_rRNA_bact_arch_phylum > 0.005]
all_rRNA_bact_arch_phylum = all_rRNA_bact_arch_phylum.dropna(axis=0, how='all')
all_rRNA_bact_arch_phylum = all_rRNA_bact_arch_phylum.fillna(0)
all_rRNA_bact_arch_phylum
```

```python
all_rRNA_eukar_phylum = (all_rRNA_eukar_phylum / all_rRNA_eukar_phylum.sum())
all_rRNA_eukar_phylum = all_rRNA_eukar_phylum[all_rRNA_eukar_phylum > 0.005]
all_rRNA_eukar_phylum = all_rRNA_eukar_phylum.dropna(axis=0, how='all')
all_rRNA_eukar_phylum = all_rRNA_eukar_phylum.fillna(0)
all_rRNA_eukar_phylum
```

```python
all_rRNA_bact_arch_phylum = (all_rRNA_bact_arch_phylum / all_rRNA_bact_arch_phylum.sum())
all_rRNA_bact_arch_phylum.round(decimals=4)*100
```

```python
all_rRNA_eukar_phylum = (all_rRNA_eukar_phylum / all_rRNA_eukar_phylum.sum())
all_rRNA_eukar_phylum.round(decimals=4)*100
```
